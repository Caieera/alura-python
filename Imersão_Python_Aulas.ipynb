```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Analise", sheet_name="Principal")
df_principal
```

```python
df_total_de_acoes = pd.read_excel("/content/Analise", sheet_name="Total_de_acoes")
df_total_de_acoes
```

```python
df_ticker = pd.read_excel("/content/Analise", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Analise", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal[["Ativo","Data","Último (R$)","Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal = df_principal[["Ativo","Data","valor_final","var_dia_pct"]].copy()
df_principal
```

```python
df_principal["Var_pct"] = df_principal['var_dia_pct']/100
df_principal["valor_inicial"] = df_principal["valor_final"] / (df_principal["Var_pct"]+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_de_acoes,left_on='Ativo',right_on='Código',how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns =['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'] )*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtd)teorica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker,left_on='Ativo',right_on='Ticker',how='left')
df_principal = df_principal.drop(columns =['Ticker'])
df_principal
```

```python
df_principal.info()
```

```python
df_principal = df_principal.merge(df_chatgpt,left_on='Nome',right_on='Nome da Empresa',how='left')
df_principal = df_principal.drop(columns =['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (em anos)':'idade_anos'}).copy()
df_principal
```

```python
df_principal['idade_anos'] = pd.to_numeric(df_principal['idade_anos'], errors='coerce').fillna(0).astype(int)
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['idade_anos'].apply(lambda x: "Mais de 100" if x > 100 else ("Menos de 50" if x < 50 else "Entre 50 e 100"))
df_principal
```

```python
# Cálculos
maior = df_principal['Variacao_rs'].max()
menor = df_principal['Variacao_rs'].min()
media = df_principal['Variacao_rs'].mean()
media_sub = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_des = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Formatando os resultados para ficar similar à sua tabela
maior = f"R$ {maior:,.2f}"
menor = f"R$ {menor:,.2f}"
media = f"R$ {media:,.2f}"
media_sub = f"R$ {media_sub:,.2f}"
media_des = f"R$ {media_des:,.2f}"

# Imprimindo os resultados
print("Maior\t", maior)
print("Menor\t", menor)
print("Media\t", media)
print("Media Sub\t", media_sub)
print("Media Des\t", media_des)
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")["Variacao_rs"].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("Resultado")["Variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y="Variacao_rs", text="Variacao_rs", title="Variação Reais por Resultado")
fig.show()
```
